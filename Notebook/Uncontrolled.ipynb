```python
import cv2
import numpy as np

image = cv2.imread('20220427_132332.png')
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
#img_hsv = hsv[:, :,2]
value = 255 #whatever value you want to add
cv2.add(hsv[:,:,2], value, hsv[:,:,2])
#image = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

cv2.imwrite('20220427_132332_hsv.png', hsv)
```

```python
img = cv2.imread('20220520_060112.png') #load rgb image
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV) #convert it to hsv

h, s, v = cv2.split(hsv)
v += 255
final_hsv = cv2.merge((h, s, v))

img = cv2.cvtColor(final_hsv, cv2.COLOR_HSV2BGR)
cv2.imwrite('20220520_060112_rev.png', img)
```
