```python
dataset_name = "open_noLED_merged"
your_dir_path = '/content/gdrive/MyDrive/Hasil_Train/Dataset_128/open_noLED_png_saturation' # String. The path of your directory containing all your subdirectories
your_dir_path2 = '/content/gdrive/MyDrive/Hasil_Train/Dataset_128/open_noLED_png_LBP_blurfirst'
new_dir_path = f'/content/{dataset_name}' # The path of your new directory with the same architecture as the previous one, but with cropped and aligned faces
```

```python
for subfolder in next(os.walk(your_dir_path))[1] : # Gives the list of all subdirectories inside the parent directory
  os.makedirs(os.path.join(new_dir_path, subfolder)) # Creates the new subdirectory. Note that it will also create new_dir_path, so there's no need to add a line os.makedirs(new_dir_path)
  if not os.path.exists(os.path.dirname(new_dir_path)):
       os.makedirs(os.path.join(new_dir_path, subfolder))
  for file in os.listdir(os.path.join(your_dir_path, subfolder)) : # Gives the list of all files inside the 'subfolder' directory
    for file in os.listdir(os.path.join(your_dir_path2, subfolder)):
      if file.endswith('.png'):    
        print("converting " + str(file))
        size = len(file)
        filename = file[:size - 4]

        image1 = Image.open(os.path.join(your_dir_path, subfolder, file))
        image2 = Image.open(os.path.join(your_dir_path2, subfolder, file))
        image = Image.blend(image1,image2,0.5)
        
        image.save(os.path.join(new_dir_path , subfolder, f'{filename}.png'))
```
