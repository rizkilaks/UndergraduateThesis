**FITTING REGRESI**

```python
from sklearn.svm import SVR
regressor = SVR(C=0.1, epsilon=0.1, kernel='linear')

regressor.fit(x_train, y_train)
```

**PREDICT & GRAPH**

```python
train_pred = regressor.predict(x_train)
train_true = y_train
train_residuals = train_true - train_pred

test_pred = regressor.predict(x_test)
test_true = y_test
test_residuals = test_true - test_pred
```

```python
from sklearn.metrics import r2_score as r2
r2_train = r2(train_true,train_pred)
r2_test = r2(test_true,test_pred)
print(format(r2_train,".3f"))
print(format(r2_test,".3f"))
```

```python
from sklearn.metrics import mean_absolute_error as MAE
MAE_train = MAE(train_true,train_pred)
MAE_test = MAE(test_true,test_pred)
print(format(MAE_train,".3f"))
print(format(MAE_test,".3f"))
```

```python
from sklearn.metrics import mean_squared_error as MSE
MSE_train = MSE(train_true,train_pred)
MSE_test = MSE(test_true,test_pred)
print(format(MSE_train,".3f"))
print(format(MSE_test,".3f"))
```
