```python
from pathlib import Path
import os.path
image_dir = Path('/content/gdrive/MyDrive/Test1/DATA/SATURATION/0deg_LED_saturation')
```

```python
import cv2
import numpy as np

images = []
for subfolder in next(os.walk(image_dir))[1]:
  for file in os.listdir(os.path.join(image_dir, subfolder)):
    #images = []
    if file.endswith('.png'):
      print("converting " + str(file))
      img = cv2.imread(os.path.join(image_dir, subfolder, file))
      img = img[:, :,1]
      img = np.array(img).flatten()
      images.append((img) / 255.0)

    #images = np.array(images)
```

```python
import pandas as pd
images = pd.Series(images)
```
